```python
# Move the training data into the correct directory
src_dir = "/content/drive/MyDrive/orca-data-for-animal-spot/Kaggle-Orca-Data"
dest_dir = "/content/drive/MyDrive/acoustic-sandbox/unpacked-detection/split/bignoise"

# Get a list of all .wav files in the source directory
wav_files = [f for f in os.listdir(src_dir) if f.endswith(".wav")]

# List of additional files to move
additional_files = ["test", "test.csv", "train", "train.csv", "val", "val.csv"]

# Move each .wav file to the destination directory
for wav_file in wav_files:
    src_path = os.path.join(src_dir, wav_file)
    dest_path = os.path.join(dest_dir, wav_file)
    shutil.move(src_path, dest_path)
    print(f"Moved {wav_file} to {dest_dir}")

# Move additional files
for additional_file in additional_files:
    src_path = os.path.join(src_dir, additional_file)
    dest_path = os.path.join(dest_dir, additional_file)
    shutil.move(src_path, dest_path)
    print(f"Moved {additional_file} to {dest_dir}")

# Move the "bkp" folder and its contents
bkp_src_path = os.path.join(src_dir, "bkp")
bkp_dest_path = os.path.join(dest_dir, "bkp")

if os.path.exists(bkp_src_path):
    try:
        shutil.copytree(bkp_src_path, bkp_dest_path)
        print(f"Moved {bkp_src_path} and its contents to {bkp_dest_path}")
    except shutil.Error as e:
        print(f"Error moving {bkp_src_path}: {e}")
    except OSError as e:
        print(f"Error moving {bkp_src_path}: {e.strerror}")
else:
    print("The 'bkp' folder does not exist in the source directory.")

print("Done!")
```

```python
!mkdir /content/drive/MyDrive/ANIMAL-SPOT/Models
def move_file(source_dir, source_file, target_dir):
    # Construct the full source file path
    source_path = os.path.join(source_dir, source_file)

    # Construct the full target file path
    target_path = os.path.join(target_dir, source_file)

    # Check if the source file exists
    if os.path.isfile(source_path):
        try:
            # Move the file to the target directory
            shutil.move(source_path, target_path)
            print(f"File '{source_file}' moved successfully from '{source_dir}' to '{target_dir}'")
        except shutil.Error as e:
            print(f"Error occurred while moving the file: {e}")
    else:
        print(f"Source file '{source_file}' not found in '{source_dir}'")

source_directory = "/content/drive/MyDrive/orca-data-for-animal-spot/Kaggle-Orca-Data/Pickle"
target_directory = "/content/drive/MyDrive/ANIMAL-SPOT/Models"
file_to_move = "Copy of ANIMAL-SPOT-FINAL.pk"

move_file(source_directory, file_to_move, target_directory)
```

```python
def rename_file(current_file_name, new_file_name):
    try:
        # Check if the current file exists
        if os.path.isfile(current_file_name):
            # Construct the new file path
            new_file_path = os.path.join(os.path.dirname(current_file_name), new_file_name)

            # Rename the file
            os.rename(current_file_name, new_file_path)
            print(f"File renamed successfully: {new_file_path}")
        else:
            print(f"Error: The file '{current_file_name}' does not exist.")
    except Exception as e:
        print(f"Error renaming file: {e}")

# Example usage
current_file_name = "/content/drive/MyDrive/ANIMAL-SPOT/Models/Copy of ANIMAL-SPOT-FINAL.pk"
new_file_name = "ANIMAL-SPOT-FINAL.pk"
rename_file(current_file_name, new_file_name)
```
